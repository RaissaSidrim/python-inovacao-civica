### Acesse a api da camara, e recupere os reembolsos do ano de 2018

```python
import requests #Importando a biblioteca 'requests', que permite que você envie solicitações HTTP em Python.

# fazendo o download do conteúdo do arquivo
url = "http://www.camara.leg.br/cotas/Ano-2019.csv.zip" #informando a URL, o ano, o formato .csv e o zip
r = requests.get(url) #fazendo uma requisição usando o método 'get'
#poderia ter colocado o link direto destro desse parêntese, mas preferi criar uma variável 'url' anteriormente
```

```python
#abrimos e escrevemos o arquivo no nosso computador
file = open('Ano-2019.csv.zip', 'wb') 
file.write(r.content)
file.close()
```

```python
#descompactamos o zip que foi baixado
from zipfile import ZipFile
zip_file = ZipFile("Ano-2019.csv.zip", 'r')
```

```python
zip_file.extract(member="Ano-2019.csv", path="reembolso-2019") #extrai o arquivo "ano-2019.csv" do zip e manda pra pasta 'reembolso-2019'
```

```python
zip_file.close()
```

### Converta a coluna `datEmissao` para `datetime`, e descubra quantos reembolsos aconteram no primeiro semestre

```python
#datatime é um tipo de dado, como str, float, int...
#ex: se uma data está dentro de um str ('03-11-1991') eu posso transformar esse string em datetime fazendo a conversão

df_reembolso['datEmissao'] = pd.to_datetime(df_reembolso['datEmissao'], format = '%Y-%m-%d')
#a primeira parte do argumento (df_reembolso['datEmissao']) é colocada para que ele faça a conversão. se eu não colocar, ele só me mostra uma visualização de como ficaria, mas sem alterar o original
#eu passei pra ele uma série de informações que ele vai processar (pd.to_datetime(df_reembolso['datEmissao'], format = ''%Y-%m-%d'')), e ele vai me devolver um dado alterado. Mas vou dizer pra ele que quero que esse dado alterado seja novamente colocado na coluna original, por isso coloco o primeiro argumento (df_reembolso['datEmissao'])
```

### Limpe a coluna `txtCNPJCPF`: ela deve conter somente dígitos numéricos

```python
# vamos trocar qualquer caracter que não seja um dígito número por '', essa operação remove os dígitos
df_reembolso["txtCNPJCPF"] = df_reembolso["txtCNPJCPF"].str.replace(r'\D', '', regex=True)

#coluna que vai receber a mudança (df_reembolso["txtCNPJCPF"]) + a alteração em si ((df_reembolso["txtCNPJCPF"].str.replace(r'\D', ''))
```
